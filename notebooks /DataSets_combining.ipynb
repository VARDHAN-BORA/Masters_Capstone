```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/202001-citibike-tripdata_1.csv")
```

```python
df2 = pd.read_csv("/content/202001-citibike-tripdata_2.csv")
```

```python
df
```

```python
df2.tail()
```

```python
df_combined = pd.concat([df, df2], ignore_index=True)
```

```python
df_combined
```

```python
# Convert 'started_at' to datetime format
df_combined['started_at'] = pd.to_datetime(df_combined['started_at'])

# Extract 15-minute intervals
df_combined['time_interval'] = df_combined['started_at'].dt.floor('15T')

# Group by start_station_id and time_interval to count the number of bookings
grouped = df_combined.groupby(['start_station_id', 'time_interval']).size().reset_index(name='booking_count')

# Create a full time range for a day (e.g., for January 1st, 2020)
time_range = pd.date_range(start='2020-01-01 00:00:00', end='2020-01-01 23:45:00', freq='15T')

# Get unique start_station_id values from the dataset
station_ids = df_combined['start_station_id'].unique()

# Create a MultiIndex for all possible combinations of start_station_id and time_interval
all_intervals = pd.MultiIndex.from_product([station_ids, time_range], names=['start_station_id', 'time_interval'])

# Reindex the grouped DataFrame to include all 15-minute intervals for each station
grouped = grouped.set_index(['start_station_id', 'time_interval']).reindex(all_intervals, fill_value=0).reset_index()
```

```python
display(grouped)
```

```python
# Group by start_station_id to get the total booking count for each station
total_bookings_per_station = grouped.groupby('start_station_id')['booking_count'].sum().reset_index()

# Optionally, display the result
total_bookings_per_station
```

```python

# Sort the DataFrame by total booking count in descending order
top_bookings = total_bookings_per_station.sort_values('booking_count', ascending=False)

# Display the top bookings
top_bookings
```

```python
# Merge top_bookings with the original DataFrame to get station names and latitude
top_bookings = pd.merge(top_bookings, df_combined[['start_station_id', 'start_station_name', 'start_lat']],
                        on='start_station_id', how='left').drop_duplicates(subset=['start_station_id'])

# Ensure 'start_station_name' and 'start_lat' are in the columns before reordering
if 'start_station_name' not in top_bookings.columns or 'start_lat' not in top_bookings.columns:
    raise ValueError("start_station_name or start_lat not found after merge. Check your data.")

# Reorder columns to place start_station_name and start_lat next to start_station_id
cols = list(top_bookings.columns)
start_station_name_index = cols.index('start_station_name')
start_lat_index = cols.index('start_lat')
cols = cols[:1] + [cols[start_station_name_index]] + [cols[start_lat_index]] + cols[1:start_station_name_index] + cols[start_station_name_index + 1:start_lat_index] + cols[start_lat_index + 1:]
top_bookings = top_bookings[cols]

# Display the updated DataFrame
top_bookings
```

# Dont forget to use different naming convention through out the files

```python

grouped.to_csv('202001_grouped.csv', index=False)
top_bookings.to_csv('202001_top_bookings.csv', index=False)

```

```python

import os

def get_file_size_gb(file_path):
  """Calculates the size of a file in gigabytes."""
  try:
    size_bytes = os.path.getsize(file_path)
    size_gb = size_bytes / (1024 ** 3)
    return size_gb
  except FileNotFoundError:
    print(f"File not found: {file_path}")
    return 0


file1_path = "/content/202001-citibike-tripdata_1.csv"
file2_path = "/content/202001-citibike-tripdata_2.csv"


size1_gb = get_file_size_gb(file1_path)
size2_gb = get_file_size_gb(file2_path)

print(f"Size of {file1_path}: {size1_gb:.2f} GB")
print(f"Size of {file2_path}: {size2_gb:.2f} GB")
```
