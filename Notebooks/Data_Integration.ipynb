```python
import pandas as pd

# âœ… Load Citi Bike Data
df_citibike = pd.read_csv('/content/drive/MyDrive/Capstone /citibike_tripdata.csv')
print("âœ… Citi Bike Data Loaded!")
df_citibike.info()
df_citibike.head(10)
```

```python
from scipy.spatial import cKDTree
import numpy as np

# âœ… Step 1: Drop conflicting columns BEFORE merging
conflicting_columns = [
    'Name', 'Latitude', 'Longitude',
    'Borough', 'Type', 'Summary', 'Full_Address', 'Neighborhood'
]
df_citibike.drop(columns=conflicting_columns, inplace=True, errors='ignore')

# âœ… Step 2: Clean data (remove NaN and infinity values)
df_citibike = df_citibike.dropna(subset=['start_lat', 'start_lng'])
df_cultural = df_cultural.dropna(subset=['Latitude', 'Longitude'])

df_citibike = df_citibike.replace([np.inf, -np.inf], np.nan).dropna(subset=['start_lat', 'start_lng'])
df_cultural = df_cultural.replace([np.inf, -np.inf], np.nan).dropna(subset=['Latitude', 'Longitude'])

# âœ… Step 3: Build KDTree for nearest neighbor search (for cultural data)
tree = cKDTree(df_cultural[['Latitude', 'Longitude']].values)

# âœ… Step 4: Find the closest cultural location for each Citi Bike station
distances, indices = tree.query(df_citibike[['start_lat', 'start_lng']].values)

# âœ… Step 5: Assign matched values
df_citibike['nearest_cultural_name'] = df_cultural.iloc[indices]['Name'].values
df_citibike['nearest_cultural_lat'] = df_cultural.iloc[indices]['Latitude'].values
df_citibike['nearest_cultural_lng'] = df_cultural.iloc[indices]['Longitude'].values
df_citibike['nearest_cultural_distance'] = distances  # (in degrees)

# âœ… Step 6: Merge using nearest coordinates
df_citibike = df_citibike.merge(
    df_cultural,
    left_on=['nearest_cultural_lat', 'nearest_cultural_lng'],
    right_on=['Latitude', 'Longitude'],
    how='left',
    suffixes=('_bike', '_cultural')
)

# âœ… Step 7: Drop unnecessary duplicate columns AFTER merge
df_citibike.drop(
    columns=[
        'Borough_bike', 'Type_bike', 'Full_Address_bike',
        'Neighborhood_bike', 'Summary_bike'
    ],
    inplace=True,
    errors='ignore'
)

# âœ… Step 8: Confirm result
print("âœ… Cultural merge completed successfully!")
df_citibike.info()
df_citibike.head(10)
```
